```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Marks DataSet.csv')
df
```

```python
df_dublicated = df
df_dublicated
```

```python
df = df.drop(columns=['Name', 'Date Of Birth', 'Age', 'Address', 'Phone Number'])
df
```

```python
df[["M1", "M2", "M3", "M4", "M5"]] = df[["M1", "M2", "M3", "M4", "M5"]].fillna(df[["M1", "M2", "M3", "M4", "M5"]].mean(numeric_only=True))
df
```

```python
X = df[['M1', 'M2', 'M3', 'M4', 'M5']]
y = df['Pass'].map({'Yes': 1, 'No': 0})

X
```

```python
y
```

```python
y.isnull()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators = 100, random_state = 42)
model.fit(X_test, y_test)

y_prediction = model.predict(X_test)
y_prediction
```
